# Notebook de Experimentación — Advertising Dataset

Este notebook contiene ejemplos de análisis y visualización del dataset *Advertising*,
utilizado en el proyecto para el modelo de regresión lineal.

Incluye:
- Carga de datos
- Exploración inicial
- Gráficos simples
- Entrenamiento de un modelo de regresión


```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("C:/Users/luish/Downloads/advertising.csv")
df.head()
```

## Visualización básica
Gráfico de la relación entre inversión en TV y ventas.


```python
plt.scatter(df["TV"], df["Sales"])
plt.xlabel("Inversión en TV")
plt.ylabel("Ventas")
plt.title("Relación entre inversión en TV y ventas")
plt.show()
```

## Entrenamiento rápido del modelo
A continuación se entrena un modelo de regresión lineal simple.


```python
X = df[["TV", "Radio", "Newspaper"]]
y = df["Sales"]

model = LinearRegression()
model.fit(X, y)

print("Coeficientes:", model.coef_)
print("Intercept:", model.intercept_)
```
